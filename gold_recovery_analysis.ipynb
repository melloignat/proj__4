# Восстановление золота из руды

Нам поставлена задача подготовить прототип модели машинного обучения для компании «Цифра»,  разрабатывающей решения для эффективной работы промышленных предприятий.

Наша модель должна предсказать коэффициент восстановления золота из золотосодержащей руды. Она поможет оптимизировать производство, чтобы не запускать предприятие с убыточными характеристиками.

Для оценки прототипа будет использоваться следующая метрика:

$$ sMAPE_{resulting} = {25\% * sMAPE_{rougher} + 75\% * sMAPE_{final} }  $$

### Проверка правильности расчёта эффективности обогащения путём расчёта средней абсолютной ошибки (MAE):

Проверим, что эффективность обогащения рассчитана правильно. Вычислим её на обучающей выборке для признака rougher.output.recovery. Найдём MAE между нашими расчётами и значением признака.

Для того, чтобы рассчитать эффективность обогащения (признак rougher.output.recovery), воспользуемся формулой:
    
$$ recovery = {C  *  (F - T) \over F  *  (C - T)} * 100\% $$

Где:

- C — доля золота в концентрате после флотации/очистки (признак rougher.output.concentrate_au)
- F — доля золота в сырье/концентрате до флотации/очистки (признак rougher.input.feed_au)
- T — доля золота в отвальных хвостах после флотации/очистки (rougher.output.tail_au)

Для вычисления MAE мы переведём в массив уже известные нам данные:

```python
# creating array of real data from the Series of the respective dataset
real_rougher_output_recovery = gold_recovery_train['rougher.output.recovery'].array
```

И выведём их на экран, просто, чтобы убедиться, что все нормально:

```python
# checking the created array to see that all is OK
real_rougher_output_recovery
```

И расчитаем нашу, расчётную величину для каждого значения признаков. Для этого из основного датасета сначала выберем то, что нам надо:

```python
# selecting from the main dataset only necessary for future calculation columns to the new small dataset
calculated_rougher_output_recovery = gold_recovery_train[['rougher.output.concentrate_au','rougher.input.feed_au',\
                                                          'rougher.output.tail_au']]
```

```python
# checking that we have all we need
calculated_rougher_output_recovery.info()
```

Далее мы рассчитываем еще раз значение эффективности обогащения по формуле выше на основании наших данных:

```python
# filtering the Pandas warning about copy of the dataset
pd.options.mode.chained_assignment = None
```

```python
# calculating the values for the new 'rougher.output.recovery' column based on data we have
calculated_rougher_output_recovery['rougher.output.recovery'] = \
(calculated_rougher_output_recovery['rougher.output.concentrate_au'] * \
 (calculated_rougher_output_recovery['rougher.input.feed_au'] - \
 calculated_rougher_output_recovery['rougher.output.tail_au'])) / \
(calculated_rougher_output_recovery['rougher.input.feed_au'] * \
(calculated_rougher_output_recovery['rougher.output.concentrate_au'] - \
calculated_rougher_output_recovery['rougher.output.tail_au'])) * 100
```

```python
# converting resulting Series to array for MAE calculation later
calculated_rougher_output_recovery = calculated_rougher_output_recovery['rougher.output.recovery'].array
```

```python
# calculating the MAE metric based on real and calculated data and formatting the output
print(format(mean_absolute_error(real_rougher_output_recovery, calculated_rougher_output_recovery), '.25f'))
```

```python
# Calculate the T-test for the means of test and train samples.
alpha = 0.05
results = st.ttest_ind(stat_feed['rougher.input.feed_size'].array, \
                       gold_recovery_train['rougher.input.feed_size'].array, equal_var=False)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

```python
# Calculate the T-test for the means of test and train samples.
alpha = 0.05
results = st.ttest_ind(stat_feed['primary_cleaner.input.feed_size'].array, \
                       gold_recovery_train['primary_cleaner.input.feed_size'].array, equal_var=False)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

### Функция для вычисления sMAPE и итоговой sMAPE

Для  оценки успешности модели подготовим функцию, которая будет помогать нам рассчитывать метрику sMAPE (англ. Symmetric Mean Absolute Percentage Error) - симметричное среднее абсолютное процентное отклонение.

$$ sMAPE = {1 \over N } \sum_{i=1}^{N} {|Y_i  -  \hat{Y_i}| \over {(|Y_i| +  \hat{|Y_i|)} / 2}} * 100\%  $$

И тогда, итоговая метрика, с учётом всех целевых признакав будет равна:

$$ sMAPE_{resulting} = {25\% * sMAPE_{rougher} + 75\% * sMAPE_{final} }  $$



```python
# function to calculate sMAPE for target feature
# A - actual data
# F - forecasted data
def smape (A, F):
    return 1/len(A) * np.sum(2 * np.abs(F-A) / (np.abs(A) + np.abs(F))*100)
```

После того, как мы рассчитаем sMAPE для выбранной модели и целевого признака, мы сможем вычислить и итоговую sMAPE по формуле, которую привели выше.
